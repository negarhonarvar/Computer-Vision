```python
import imutils
!pip install easyocr
import easyocr
from skimage import io, color
import imageio as iio
from PIL import Image, ImageOps
```

```python
image_61 = cv2.imread(r"/content/drive/MyDrive/Colab Notebooks/FCV-HW2/image6.1.jpg")
gray_61 = cv2.cvtColor(image_61, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_61,cmap = 'gray')

image_62 = cv2.imread(r"/content/drive/MyDrive/Colab Notebooks/FCV-HW2/image6.2.jpg")
gray_62 = cv2.cvtColor(image_62, cv2.COLOR_BGR2GRAY)
# plt.imshow(gray_62,cmap = 'gray')
```

```python
filtered_image_61 = cv2.Canny(cv2.bilateralFilter(gray_61, 11, 17, 17), 30, 200)
keypoints_61 = cv2.findContours(filtered_image_61.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
contours_61 = sorted(imutils.grab_contours(keypoints_61), key=cv2.contourArea,reverse=True)

filtered_image_62 = cv2.Canny(cv2.bilateralFilter(gray_62, 11, 17, 17), 30, 200)
keypoints_62 = cv2.findContours(filtered_image_62.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
contours_62 = sorted(imutils.grab_contours(keypoints_62), key=cv2.contourArea,reverse=True)
```

```python
location_61 = None
for contour in contours_61:
    approx = cv2.approxPolyDP(contour, 10, True)
    if len(approx) == 4:
      location_61 = approx
      break
location_62 = None
for contour in contours_62:
    approx = cv2.approxPolyDP(contour, 10, True)
    if len(approx) == 4:
      location_62 = approx
      break
```

```python
mask_61 = np.zeros(gray_61.shape, np.uint8)
new_image_61= cv2.drawContours(mask_61, [location_61], 0, 255, -1)
new_image_61 =  cv2.bitwise_and(image_61, image_61, mask=mask_61)

mask_62 = np.zeros(gray_62.shape, np.uint8)
new_image_62= cv2.drawContours(mask_62, [location_62], 0, 255, -1)
new_image_62 =  cv2.bitwise_and(image_62, image_62, mask=mask_62)
```

```python
plt.imshow(cv2.cvtColor(new_image_62, cv2.COLOR_BGR2RGB))
```

```python
(x, y) = np.where(mask_62 ==255)
(x1, y1) = (np.min(x), np.min(y))
(x2, y2) = (np.max(x), np.max(y))
cropped_image = gray_62[x1:x2+1, y1:y2+1]
reader = easyocr.Reader(['en'])
result = reader.readtext(cropped_image)
result
```
